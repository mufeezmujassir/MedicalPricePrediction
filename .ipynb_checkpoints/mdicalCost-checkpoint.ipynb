```python
x=medical.drop('charges', axis=1)
y=medical['charges']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestRegressor
```

```python
ohe=OneHotEncoder()
ohe.fit(x[['sex','smoker','region']])

column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['sex','smoker','region']),remainder='passthrough')
print(column_trans)
```

```python
model=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,model)
```

```python
pipe.fit(x_train,y_train)
```

```python
pred=pipe.predict(x_test)
pred
```

```python
r2_score(y_test,pred)
```

```python
scores=[]
for i in range(1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=i,test_size=0.2)
    lr=LinearRegression()
    model=make_pipeline(column_trans,lr)
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    scores.append(r2_score(y_test,y_pred))

print(scores[np.argmax(scores)])
print(np.argmax(scores))
```

 # create the model

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=np.argmax(scores),test_size=0.2)
lr=LinearRegression()
model=make_pipeline(column_trans,lr)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
r2_score(y_test,y_pred)
```

```python
score=[]
for i in range(10,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=i,test_size=0.2)
    RF = RandomForestRegressor(n_estimators=i)
    model = make_pipeline(column_trans, RF)
    model.fit(x_train, y_train)
    pred = model.predict(x_test)
    score.append(r2_score(y_test, pred))
print(score[np.argmax(score)])
print(np.argmax(score))
```

```python
RF = RandomForestRegressor(n_estimators=57)
model = make_pipeline(column_trans, RF)
model.fit(x_train, y_train)
pred = model.predict(x_test)
score.append(r2_score(y_test, pred))
```
